Let's begin with understanding probability distributions, focusing on the normal distribution. 

The normal distribution is foundational in statistics and economics. Its key properties:
1. Symmetric bell shape
2. Defined by mean (μ) and standard deviation (σ)
3. ~68% of data within 1σ, ~95% within 2σ, ~99.7% within 3σ

Let's start by visualizing this:

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create x-values for our plot
x = np.linspace(-4, 4, 1000)

# Calculate normal distribution values
y = norm.pdf(x, 0, 1)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', label='Standard Normal')
plt.title('Standard Normal Distribution')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

````python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create x-values for our plot
x = np.linspace(-4, 4, 1000)

# Calculate normal distribution values
y = norm.pdf(x, 1, 1.2)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', label='Standard Normal')
plt.title('Standard Normal Distribution')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
````
